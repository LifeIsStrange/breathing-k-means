## Symbols used

* $\phi$ : Summed Square Error (i.e., SSE)

* $t$ : CPU time

* $\Delta\phi = 1 - \phi_{\textrm{bkm}}/\phi_{\textrm{km++}}$ (positive values indicate better solutions from `breathing k-means`)

* $\Delta t = 1 - t_{\textrm{bkm}}/t_{\textrm{km++}}$ (positive values indicate faster computation by `breathing k-means`)
